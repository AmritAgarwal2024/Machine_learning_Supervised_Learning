## Data Description




A dataset of Imports-Exports is taken from kaggle.com https://www.kaggle.com/datasets/chakilamvishwas/imports-exports-15000) with 15000 rows x 16 columns,contains data related to international trade transactions. Each row represents a single transaction with the following columns:

Columns:

* Transaction_ID: A unique identifier for each transaction.

* Country: The country involved in the transaction, either as the origin or destination.

* Product: The type of product being traded.

* Import_Export: Indicates whether the transaction is an import or export.

* Quantity: The quantity of the product traded.

* Value: The total value of the transaction in dollars.

* Date: The date the transaction occurred, in the format DD-MM-YYYY.

* Category: The category of the traded product, such as Machinery, Clothing, Electronics, Furniture, or Toys.

* Port: The port where the transaction took place.

* Customs_Code: A unique code assigned by customs authorities for the transaction.

* Weight: The total weight of the traded goods in kilograms.

* Shipping_Method: The method of shipping used for the transaction, either Air or Sea.

* Supplier: The company supplying the traded goods.

* Customer: The customer receiving the traded goods.

* Invoice_Number: A unique identifier for the invoice related to the transaction.

* Payment_Terms: The payment terms agreed upon for the transaction, such as Cash on Delivery, Prepaid, Net 30, Net 60, etc.

The data covers transactions from various countries and includes both imports and exports across different product categories. The transactions span from 2019 to 2024, with the majority occurring in 2021 and 2022.g in 2021 and 2022.

```python

# Create an empty dictionary to store the results
dispersion_results = {}

# Loop through each column in the DataFrame
for column in aa04_sdf_noncat.columns:
    # Calculate measures of dispersion
    data = aa04_sdf_noncat[column].dropna()  # Remove missing values for calculations
    range_val = data.max() - data.min()
    std_dev = data.std()
    skewness = skew(data)
    kurtosis_val = kurtosis(data)

    # Store the results in the dictionary
    dispersion_results[column] = {
        'Range': range_val,
        'Standard Deviation': std_dev,
        'Skewness': skewness,
        'Kurtosis': kurtosis_val
    }

# Display the results
for column, stats in dispersion_results.items():
    print(f"Variable: {column}")
    for stat_name, stat_value in stats.items():
        print(f"  {stat_name}: {stat_value}")
    print("-" * 20)
```

```python
# Pearson and Spearman correlation

# Create an empty DataFrame to store the results
correlation_results = pd.DataFrame(index=aa04_sdf_noncat.columns,
                                    columns=['Pearson', 'Spearman'])

# Loop through each pair of columns in the DataFrame
for col1 in aa04_sdf_noncat.columns:
    for col2 in aa04_sdf_noncat.columns:
        if col1 != col2:  # Avoid calculating correlation with itself
            # Calculate Pearson correlation
            pearson_corr, _ = pearsonr(aa04_sdf_noncat[col1],
                                       aa04_sdf_noncat[col2])

            # Calculate Spearman correlation
            spearman_corr, _ = spearmanr(aa04_sdf_noncat[col1],
                                         aa04_sdf_noncat[col2],
                                         nan_policy='omit')  # Handle missing values

            # Store the results in the DataFrame
            correlation_results.loc[col1, 'Pearson'] = pearson_corr
            correlation_results.loc[col1, 'Spearman'] = spearman_corr

# Display the results
print("Correlation Results:\n", correlation_results)
```

```python
# coefficient of variation and confidence interval

# Create an empty dictionary to store the results
composite_results = {}

# Loop through each column in the DataFrame
for column in aa04_sdf_noncat.columns:
    # Calculate measures
    data = aa04_sdf_noncat[column].dropna()  # Remove missing values for calculations

    # Coefficient of Variation
    cv = (data.std() / data.mean()) * 100

    # Confidence Interval (95% confidence level)
    confidence_level = 0.95
    degrees_of_freedom = len(data) - 1

    # Calculate standard error of the mean
    sem_val = sem(data)

    # Calculate confidence interval
    confidence_interval = t.interval(confidence_level,
                                     degrees_of_freedom,
                                     loc=data.mean(),
                                     scale=sem_val)

    # Store the results in the dictionary
    composite_results[column] = {
        'Coefficient of Variation': cv,
        'Confidence Interval (95%)': confidence_interval
    }

# Display the results
for column, stats in composite_results.items():
    print(f"Variable: {column}")
    for stat_name, stat_value in stats.items():
        print(f"  {stat_name}: {stat_value}")
    print("-" * 20)
```

## Inferetial Statistics

### Categorical Inferential Statitics

```python
# Assuming aa04_sdf is your DataFrame with the relevant columns
aa04_sdf_cat = aa04_sdf[['Import_Export', 'Category', 'Shipping_Method', 'Payment_Terms']]

# Function to run Chi-square test for homogeneity
def chi_square_test_of_homogeneity(df, columns):
    results = {}
    for i in range(len(columns)):
        for j in range(i + 1, len(columns)):
            col1 = columns[i]
            col2 = columns[j]
            contingency_table = pd.crosstab(df[col1], df[col2])

            # Perform Chi-square test
            chi2, p, dof, expected = chi2_contingency(contingency_table)

            results[f"{col1} vs {col2}"] = {
                "Chi2": chi2,
                "p-value": p,
                "Degrees of Freedom": dof,
                "Expected Frequencies": expected
            }
    return results

# Example usage
chi_square_results = chi_square_test_of_homogeneity(aa04_sdf_cat, aa04_sdf_cat.columns)

# Print results
for comparison, stats in chi_square_results.items():
    print(f"\n{comparison}:")
    for key, value in stats.items():
        if key == "Expected Frequencies":
            print(f"{key}:\n{value}")
        else:
            print(f"{key}: {value}")
```

### Non Categorical Inferential Statistics

```python
import pandas as pd
import numpy as np
from scipy import stats

# Assuming 'aa04_sdf_noncat' is already defined as a DataFrame with non-categorical columns

alpha = 0.05  # Significance level

# Loop through each column in the DataFrame
for column in aa04_sdf_noncat.columns:
    data = aa04_sdf_noncat[column].dropna()  # Remove missing values for calculations

    print(f"Variable: {column}")

    # Shapiro-Wilk Test
    shapiro_stat, shapiro_p = stats.shapiro(data)
    print(f"  Shapiro-Wilk Test: statistic={shapiro_stat:.3f}, p-value={shapiro_p:.3f}, ", end="")
    if shapiro_p > alpha:
        print("Data is normally distributed (fail to reject H0)")
    else:
        print("Data is not normally distributed (reject H0)")

    # Kolmogorov-Smirnov Test
    ks_stat, ks_p = stats.kstest(data, 'norm')
    print(f"  Kolmogorov-Smirnov Test: statistic={ks_stat:.3f}, p-value={ks_p:.3f}, ", end="")
    if ks_p > alpha:
        print("Data is normally distributed (fail to reject H0)")
    else:
        print("Data is not normally distributed (reject H0)")

    # Anderson-Darling Test
    anderson_result = stats.anderson(data, dist='norm')
    print(f"  Anderson-Darling Test: statistic={anderson_result.statistic:.3f}, ", end="")
    if anderson_result.statistic < anderson_result.critical_values[2]:  # Using critical value for alpha=0.05
        print("Data is normally distributed (fail to reject H0)")
    else:
        print("Data is not normally distributed (reject H0)")

    # Jarque-Bera Test
    jb_stat, jb_p = stats.jarque_bera(data)
    print(f"  Jarque-Bera Test: statistic={jb_stat:.3f}, p-value={jb_p:.3f}, ", end="")
    if jb_p > alpha:
        print("Data is normally distributed (fail to reject H0)")
    else:
        print("Data is not normally distributed (reject H0)")

    print("-" * 20)
```
